```python
df2['Family_size']=df2['SibSp']+df2['Parch']+1 # we have calculated the family size 
df2 # As this is the numerical column, so we will convert this into categrical type 
```

```python
# Family type 
# 1-> Alone
# 2-4-> Small
# >5 -> Large

def transform_family_size(num):
    if num==1:
        return "Alone"
    elif num>1 and num<5:
        return "Small Family"
    else:
        return "Large Family"
```

```python
df2['Family_type']=df2['Family_size'].apply(transform_family_size)
```
